# Feature Extraction
### Generating new features from already existing features.

```python
# WeekDays
from data_manipulator import DataManipulator
manu = DataManipulator(train_store)
```

```python
# weekdays
manu.add_week_day('DayOfWeek')
```

```python
# number of days to holydays
manu.add_number_of_days_to_holiday('StateHoliday')
```

```python
# number of days after holydays
manu.add_number_of_days_after_holiday('StateHoliday')
```

```python
# Beginning of month, mid month and ending of month
manu.add_month_timing('Day')
```

### More Features

```python
# Adding salespercustomer column
train_store['SalesPerCustomer']=train_store['Sales']/train_store['Customers']
```

```python
# Since the competition variables CompetitionOpenSinceYear and CompeitionOpenSinceMonth have the same 
# underlying meaning, merging them into one variable
train_store['CompetitionOpenSince'] = np.where((train_store['CompetitionOpenSinceMonth']==0) & 
                                                  (train_store['CompetitionOpenSinceYear']==0) , 0,
                                                  (train_store.Month - train_store.CompetitionOpenSinceMonth) 
                                                  + (12 * (train_store.Year - train_store.CompetitionOpenSinceYear)))
```

```python
# Adding Season
manu.add_season('Month')
```

### Saving version 2 for decoded / unlabeled data

```python
train_store.to_csv('../data/train_store.csv', index=False)
```

## Labeling Object Columns

```python
train_store['StateHoliday'] = train_store['StateHoliday'].apply(lambda x: 'o' if x == '0' else x)
train_store['StateHoliday'] = train_store['StateHoliday'].apply(lambda x: 'o' if x == 0 else x)
```

```python
# Converting catagorical variables with label
manu.label_columns(['Season', 'StoreType', 'Assortment', 'StateHoliday', 'PromoInterval'])
```

## Conversion of types

```python
train_store['CompetitionOpenSinceMonth'] = train_store['CompetitionOpenSinceMonth'].astype(int)
```

```python
train_store['CompetitionOpenSinceYear'] = train_store['CompetitionOpenSinceYear'].astype(int)
```

```python
train_store['Promo2SinceWeek'] = train_store['Promo2SinceWeek'].astype(int)
```

```python
train_store['Promo2SinceYear'] = train_store['Promo2SinceYear'].astype(int)
```

```python
train_store['CompetitionOpenSince'] = train_store['CompetitionOpenSince'].astype(int)
```
